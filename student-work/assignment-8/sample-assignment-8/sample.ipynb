# Hypotheses Formulation

Based on the bar plot above, I formulated the following hypotheses:

### Hypothesis 1:
- Students with lower average final grades have failed more classes.

### Hypothesis 2:
- Students whose parents are separated have more absences on average.


# Testing Hypothesis 1:
### Hypothesis 1:
- Students with lower average final grades have failed more classes.

```python
#Testing hypothesis 1

import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for failures vs. final grade 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='failures', y='G3', alpha=0.6, color='blue')
plt.title('Figure 1: Scatter Plot of Failures vs. Final Grade ')
plt.xlabel('Number of Failures')
plt.ylabel('Final Grade ')
plt.grid()
plt.show()
```

```python
#Testing hypothesis 1

import pandas as pd
from scipy.stats import spearmanr, ttest_ind

# Assuming df is your DataFrame
# Spearman correlation
correlation, p_value = spearmanr(df['failures'], df['G3'])
print(f'Spearman Correlation: {correlation}, p-value: {p_value}')

# T-test: Compare means of grades based on failures (0 failures vs. >0 failures)
group_0_failures = df[df['failures'] == 0]['G3']
group_1_failures = df[df['failures'] > 0]['G3']
t_stat, p_value_ttest = ttest_ind(group_0_failures, group_1_failures, equal_var=False)
print(f'T-test Statistic: {t_stat}, p-value: {p_value_ttest}')

```

# Results and Analysis of Hypothesis Test 1

### Hypotheses

- **Null Hypothesis (H0):** There is no relationship between the number of failures and final grades.  
  - The Spearman correlation coefficient $ \rho = 0 $.  
  - The average final grades for students with 0 failures is equal to the average final grades for students with more than 0 failures.

- **Alternative Hypothesis (HA):** There is a negative relationship between the number of failures and final grades.  
  - The Spearman correlation coefficient $ \rho < 0 $.  
  - The average final grades for students with 0 failures is higher than the average final grades for students with more than 0 failures.

### Results

- **Scatter Plot:**
  - The downard sloping trend between grades and number of class failed in Figure 1 indicates a moderatly strong negative correlation.

- **Spearman Correlation:** -0.361223517700282  
  - This negative correlation indicates a moderate inverse relationship between the number of failures and final grades, meaning that as the number of failures increases, the average final grade tends to decrease.

- **Spearman p-value:** $ 1.28 \times 10^{-13} $  
  - The p-value is much less than a typical alpha level of 0.05 (95% confidence), allowing us to reject the null hypothesis. There is significant evidence to suggest a negative correlation between the number of failures and final grades.

- **T-test Statistic:** 7.015559311278065  
  - This T-statistic is significantly larger than the typical critical value of 2 (95% confidence), placing it in the rejection region, so we reject H0 and accept HA.

- **T-test p-value:** $ 1.53 \times 10^{-10} $  
  - This p-value is also much less than 0.05 (95% confidence), so we reject the null hypothesis. This suggests that students with 0 failures have significantly higher average final grades compared to those who have failed at least one class.

### Hypothesis 1 Conclusion

Based on the results of both the Spearman correlation and the T-test:
- There is a significant negative correlation between the number of failures and average final grades.
- Students with lower average final grades are more likely to have failed classes, supporting the alternative hypothesis (HA).


# Testing Hypothesis 2:
### Hypothesis 2:
- Students whose parents are separated have more absences on average.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Average absences of students whose parents are together vs students whose parents are seperated
plt.figure(figsize=(10, 6))
sns.barplot(x='Pstatus', y='absences', data=df, errorbar=None)  # 'Pstatus' is assumed to be the column for parental status
plt.title("Figure 2: Average Absences by Parental Status")
plt.xlabel("Parental Status (T = Together, A = Apart)")
plt.ylabel("Average Absences")
plt.show()
```

```python
#t- test for hypothesis 2

from scipy.stats import ttest_ind

together_absences = df[df['Pstatus'] == 'T']['absences']  
apart_absences = df[df['Pstatus'] == 'A']['absences'] 

t_stat, p_value = ttest_ind(apart_absences, together_absences, equal_var=False) 

print("T-test Statistic:", t_stat)
print("p-value:", p_value)
```

# Results and Analysis of Hypothesis Test 2

### Hypotheses

- **Null Hypothesis (H0):** There is no difference in average absences between students whose parents are separated and those whose parents are together.  
  - The average absences for students with separated parents is equal to the average absences for students with parents together.

- **Alternative Hypothesis (HA):** Students whose parents are separated have more absences on average.  
  - The average absences for students with separated parents is greater than the average absences for students with parents together.

### Results

- **Bar Plot Visualization:**    
  - Figure 2 shows the average absences for students with separated parents compared to those with parents together. There is a higher average for students with seperated parents supporting the alternative hypothesis. 

- **T-test Statistic:** 1.7525502919959017  
  - This T-statistic is slighter smaller than the typical critical value of 2 (95% confidence), placing in the within the fail to reject region, so we fail to reject H0.

- **T-test p-value:** 0.0868007482605107 
  -  The p-value is greater than the typical alpha level of 0.05, suggesting that we do not have enough evidence to reject the null hypothesis. 


### Hypothesis 2 Conclusion

Based on the results of the T-test:
- The T-test statistic of 1.75 indicates that while there is a difference in average absences, it is not statistically significant at the 0.05 level.
- The p-value of 0.0868 suggests that we fail to reject the null hypothesis, meaning we do not have sufficient evidence to support the claim that students whose parents are separated have more absences on average.
- Therefore, the data does not support the alternative hypothesis (HA).
